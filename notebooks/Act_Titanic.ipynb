```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

# Upper Class data

```python
df_u=df[(df['Pclass']==1)]
```

```python
#Generar una paleta de colores que es un degradado de verdes, mientras mas alto el numero, mas claro y viceversa.
c=sns.cubehelix_palette(start=2, rot=0, dark=0, light=.95, reverse=True, as_cmap=True) #https://seaborn.pydata.org/tutorial/color_palettes.html
ax = sns.histplot(data=df_u, x='Age', color=c(100))

# Agregar etiquetas con valores a cada barra
for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{height}', (p.get_x() + p.get_width() / 2., height),
                ha='center', va='baseline', fontsize=10, color='black', xytext=(0, 5),
                textcoords='offset points')

plt.show()
```

##### El rango de edades más frecuente entre los pasajeros de la clase alta va de los 40 hasta los 50 años.

```python
ax = sns.histplot(data=df_u, x='Sex', color='#b4bd84')

for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{height}', (p.get_x() + p.get_width() / 2., height),
                ha='center', va='baseline', fontsize=10, color='black', xytext=(0, 5),
                textcoords='offset points')
plt.show()
```

##### Hay más hombres que mujeres

```python
#Al hacer un histograma, los valores de x me daban en decimales, y mejor opté por usar un countplot.

ax = sns.countplot(data=df_u, x='SibSp', color='#307a2b')

for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{height}', (p.get_x() + p.get_width() / 2., height),
                ha='center', va='baseline', fontsize=10, color='black', xytext=(0, 5),
                textcoords='offset points')
plt.show()
```
